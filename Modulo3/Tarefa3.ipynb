```python
# 1. Classe Bola: Crie uma classe que modele uma bola:
# a. Atributos: Cor, circunferência, material

class Bola:
    def __init__(self, cor, circunferencia, material):
        self.cor = cor
        self.circunferencia = circunferencia
        self.material = material
        
# b. Métodos: trocaCor e mostraCor

    def trocaCor(self, novaCor):
        self.cor = novaCor
        print(f"A nova cor da bola é: {self.cor}")
        
    def mostraCor(self):
        print(f"A cor da bola é: {self.cor}")
        

bola = Bola('azul', 10, 'borracha')
bola.mostraCor()

bola.trocaCor('vermelha')
```

```python
# 2. Classe Quadrado: Crie uma classe que modele um quadrado:
# a. Atributos: Tamanho do lado

class Quadrado:
    def __init__(self, lado):
        self.lado = lado

# b. Métodos: Mudar valor do Lado, Retornar valor do Lado e calcular Área;

    def mudarLado(self, novoLado):
        self.lado = novoLado
        print(f"O novo valor do lado é: {self.lado}")
        
    def obterLado(self):
        print(f"O valor do lado é: {self.lado}")
        
    def calcularArea(self):
        area = self.lado * self.lado
        print(f"O valor da área do quadrado é: {area}")
        
quadrado = Quadrado(2)
quadrado.obterLado()
quadrado.calcularArea()

quadrado.mudarLado(5)
quadrado.obterLado()
quadrado.calcularArea()
```

```python
# 3. Classe Retangulo: Crie uma classe que modele um retangulo:
# a. Atributos: LadoA, LadoB (ou Comprimento e Largura, ou comprimento e largura, a
# escolher)4

class Retangulo:
    def __init__(self, comprimento, largura):
        self.comprimento = comprimento
        self.largura = largura

# b. Métodos: Mudar valor dos lados, Retornar valor dos lados, calcular Área e calcular
# Perímetro;

    def mudarValores(self, novoComprimento, novaLargura):
        self.comprimento = novoComprimento
        self.largura = novaLargura
        print(f"O novo valor da comprimento é: {self.comprimento} e o valor da largura é: {self.largura}")
    
    def obterValores(self):
        print(f"O valor da comprimento é: {self.comprimento} e o valor da largura é: {self.largura}")
        
    def calcularArea(self):
        area = self.comprimento * self.largura
        return area
    
    def calcularPerimetro(self):
        perimetro = (2 * self.comprimento) + (2 * self.largura)
        return perimetro

# c. Crie um programa que utilize esta classe. Ele deve pedir ao usuário que informe
# as medidas de um local. Depois, deve criar um objeto com as medidas e
# calcular a quantidade de pisos e de rodapés necessárias para o local.

comprimento = int(input("Digite o comprimento do local: "))
largura = int(input("Digite a largura do local: "))

medidasLocal = Retangulo(comprimento, largura)

print(f"Serão necessários {medidasLocal.calcularArea()} metros quadrados de pisos.")
print(f"Serão necessários {medidasLocal.calcularPerimetro()} metros de rodapés.")
```

```python
# 4. Classe Pessoa: Crie uma classe que modele uma pessoa:
# a. Atributos: nome, idade, peso e altura

class Pessoa:
    def __init__(self, nome, idade, peso, altura):
        self.nome = nome
        self.idade = idade
        self.peso = peso
        self.altura = altura
    
# b. Métodos: Envelhecer, engordar, emagrecer, crescer. Obs: Por padrão, a cada
# ano que nossa pessoa envelhece, sendo a idade dela menor que 21 anos, ela
# deve crescer 0,5 cm.

    def envelhecer(self, anos):
        idade = self.idade + anos
        return idade
    
    def engordar(self, quilos):
        novoPeso = self.peso + quilos
        return novoPeso
        
    def emagrecer(self, quilos):
        novoPeso = self.peso - quilos
        return novoPeso
    
    def crescer(self):
        if (self.idade < 21): 
            self.altura += 0.5
        return self.idade

pessoa = Pessoa('Joaquim', 7, 30, 1.1)
```

```python
# 6. Classe Conta Corrente: Crie uma classe para implementar uma conta corrente. A
# classe deve possuir os seguintes atributos: número da conta, nome do correntista e
# saldo. Os métodos são os seguintes: alterarNome, depósito e saque; No construtor,
# saldo é opcional, com valor default zero e os demais atributos são obrigatórios.

class ContaCorrente:
    def __init__(self, numeroConta, nomeTitular, saldo = 0):
        self.numeroConta = numeroConta
        self.nomeTitular = nomeTitular
        self.saldo = saldo
    
    def alterarNome(self, novoNome):
        self.nomeTitular = novoNome
        print(f"O novo nome do titular da conta corrente é: {self.nomeTitular}")

    def deposito(self, valorDeposito):
        self.saldo += valorDeposito
        print(f"O novo valor do saldo é: {self.saldo}")

    def saque(self, valorSaque):
        self.saldo -= valorSaque
        print(f"O novo valor do saldo é: {self.saldo}")


conta = ContaCorrente(123, 'Maria')

conta.alterarNome('José')

conta.deposito(10)

conta.saque(3)
```

```python
# 7. Classe TV: Faça um programa que simule um televisor criando-o como um objeto. O
# usuário deve ser capaz de informar o número do canal e aumentar ou diminuir o volume.
# Certifique-se de que o número do canal e o nível do volume permanecem dentro de
# faixas válidas.

class TV:
    def __init__(self, numeroCanal, volume):
        self.numeroCanal = numeroCanal
        self.volume = volume

    def alterarCanal(self, novoCanal):
        if (1 < novoCanal < 200):
            self.numeroCanal = novoCanal
            print(f"O novo canal é: {self.numeroCanal}")
        else: 
            print(f"Escolha um novo canal. O canal é limitado entre 1 e 200")

    def alterarVolume(self, novoVolume):
        if (0 < novoVolume < 100):
            self.novoVolume = novoVolume
            print(f"O novo volume é: {self.novoVolume}")
        else: 
            print(f"Escolha um novo volume. O volume é limitado entre 0 e 100")

tv = TV(44, 25)

tv.alterarCanal(350)
tv.alterarCanal(57)
tv.alterarVolume(102)
tv.alterarVolume(30)
```

```python
# 8. Classe Bomba de Combustível: Faça um programa completo utilizando classes e
# métodos que:
# a. Possua uma classe chamada bombaCombustível, com no mínimo esses atributos:
# i. tipoCombustivel.
# ii. valorLitro
# iii. quantidadeCombustivel

class bombaCombustível:
    def __init__(self, tipoCombustivel, valorLitro, quantidadeCombustivel):
        self.tipoCombustivel = tipoCombustivel
        self.valorLitro = valorLitro
        self.quantidadeCombustivel = quantidadeCombustivel
    
# b. Possua no mínimo esses métodos:
# i. abastecerPorValor( ) – método onde é informado o valor a ser abastecido e
# mostra a quantidade de litros que foi colocada no veículo

    def abastecerPorValor(self, valorAbastecido):
        quantidadeLitros = valorAbastecido / self.valorLitro
        print(f"Foram abastecidos {quantidadeLitros} litros de combustível")
        
# ii. abastecerPorLitro( ) – método onde é informado a quantidade em litros de
# combustível e mostra o valor a ser pago pelo cliente.

    def abastecerPorLitro(self, quantidadeLitros):
        valorAbastecimento = quantidadeLitros * self.valorLitro
        print(f"O valor a ser pago é R${valorAbastecimento}")

# iii. alterarValor( ) – altera o valor do litro do combustível.

    def alterarValor(self, novoValorLitro):
        self.valorLitro = novoValorLitro
        print(f"O novo valor do litro de combustível é R${self.valorLitro}")

# iv. alterarCombustivel( ) – altera o tipo do combustível.

    def alterarCombustivel(self, novoTipoCombustivel):
        self.tipoCombustivel = novoTipoCombustivel
        print(f"O novo tipo de combustível é: {self.tipoCombustivel}")

# v. alterarQuantidadeCombustivel( ) – altera a quantidade de combustível
# restante na bomba.

    def alterarQuantidadeCombustivel(self, novaQuantidadeCombustivel):
        self.quantidadeCombustivel = novaQuantidadeCombustivel
        print(f"A nova quantidade de combustível é: {self.quantidadeCombustivel}")
        

bomba = bombaCombustível('diesel', 3, 450)

bomba.abastecerPorValor(30)

bomba.abastecerPorLitro(15)

bomba.alterarValor(5)

bomba.alterarCombustivel('gasolina')

bomba.alterarQuantidadeCombustivel(400)
```
