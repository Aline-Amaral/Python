```python
# 1. Faça um programa que leia um arquivo texto contendo uma lista de endereços IP e gere
# um outro arquivo, contendo um relatório dos endereços IP válidos e inválidos.
# • O arquivo de entrada possui o seguinte formato:
# 200.135.80.9
# 192.168.1.1
# 8.35.67.74
# 257.32.4.5
# 85.345.1.2
# 1.2.3.4
# 9.8.234.5
# 192.168.0.256
# • O arquivo de saída possui o seguinte formato:
# [Endereços válidos:]
# 200.135.80.9
# 192.168.1.1
# 8.35.67.74
# 1.2.3.4
# [Endereços inválidos:]
# 257.32.4.5
# 85.345.1.2
# 9.8.234.5
# 192.168.0.256

import socket

entrada = open('entrada.txt', 'r')
saida = open('saida.txt', 'w')

validos = ['[Endereços válidos:]\n']
invalidos = ['[Endereços inválidos:]\n']


for line in entrada:
    try:
        socket.inet_aton(line)
        validos.append(line)
    except socket.error:
        invalidos.append(line)

saida.writelines(validos)
saida.writelines(invalidos)

entrada.close()
saida.close()
```

```python
# 2. Quando falamos sobre erros e exceções, algumas fontes de informação consideram
# que no processo de depuração, podemos encontrar três tipos de erro: erro de sintaxe,
# erro de tempo de execução e erro semântico. Defina o conceito de cada um deles
# a) Erro de sintaxe
Quando há algum erro na estrutura do código, ou seja, alguma quebra as regras ou estruturas de como o programa deve ser escrito para ser interpretado
A mensagem de erro é exibida e o programa não roda. 

# b) Erro de tempo de execução
Também chamados de exceções os erros em tempo de execução são exibidos quando o programa é executado, possuem diferentes tipos que são exibidos na mensagem de erro.

# c) Erro semântico
O programa chega a ser executado e não ocorrem exceções, porém a saída apresentada não é a esperada. O programa está escrito de acordo com as regras da linguagem, porém não retorna o que é necessário.
```

```python
# 3. Para explorar os erros e as exceções em Python, convido vocês a revisitarem os
# exercícios do primeiro módulo e para cada um deles, provocar alguns erros e avaliar as
# mensagens do interpretador. Nesse estágio do nosso curso, deve estar mais claro e
# “menos aterrorizante” perceber que o Python nos ajuda a entender alguns dos erros que
# cometemos. Podemos implementar métodos de tratamento de exceções para as
# respectivas entradas de dados em cada um dos exercícios propostos. Você pode
# considerar a utilização dos seguintes recursos:
# a) Tente primeiramente utilizar os blocos try e except tratando as entradas de dados. Por
# exemplo: se o programa pede para que entre uma string, caso o usuário entre com
# um número inteiro, uma mensagem de erro deve aparecer e indicar o usuário o
# caminho correto.
# b) Tente tratar as exceções com mensagens especificas. Dessa forma, melhoramos a
# usabilidade do código.
# c) Tente utilizar quando possível, a construção completa envolvendo try, except, else e
# finally

try: 
    a = int(input("Digite o primeiro número: "))
except: 
    print("Letras não são válidas. Digite um número")
    a = int(input("Digite o primeiro número: "))
finally:
    print(f"O primeiro número é: {a}")
try: 
    b = int(input("Digite o segundo número: "))
except: 
    print("Letras não são válidas. Digite um número")
    b = int(input("Digite o segundo número: "))
finally:
    print(f"O segundo número é: {b}")
try: 
    operacao = input("Digite o tipo de operação que deseja realizar: ")
except: 
    print("Essa entrada não é válida. Digite a operação por extenso.")
    operacao = input("Digite o tipo de operação que deseja realizar: ")
finally:
    print(f"A operação é: {operacao}")

if operacao == 'soma':
    result = a + b
elif operacao == 'subtração':
    result = a - b
elif operacao == 'multiplicação':
    result = a * b
elif operacao == 'divisão':
    result = a / b
print(f"O resultado da {operacao} dos dígitos {a} e {b} é: {result}")
```
